## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_reduced_df = application_df.drop(['EIN', 'NAME'], axis=1)
application_reduced_df.head()
```

```python
# Shape of the dataframe to ascertain number of records and features
application_reduced_df.shape
```

```python
# Determine the number of unique values in each column.
application_reduced_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_unique_counts = application_reduced_df['APPLICATION_TYPE'].value_counts()
app_type_unique_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace =  list(app_type_unique_counts[app_type_unique_counts < 528].index)
print(application_types_to_replace)

# Replace in dataframe
for app in application_types_to_replace:
    application_reduced_df['APPLICATION_TYPE'] = application_reduced_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_reduced_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_reduced_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts[classification_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(classification_counts[classification_counts < 1883].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_reduced_df.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_final_df = pd.get_dummies(application_reduced_df)
application_final_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_final_df.drop('IS_SUCCESSFUL', axis = 1).values
y = application_final_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.


nn = Sequential()

# Input layer and hidden layer
nn.add(Dense(units=108, activation="relu", input_dim=108))
nn.add(Dense(units=108, activation="relu"))

# Output layer
nn.add(Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
```

```python
# Creating a custom class to save weights after every 5 epochs

from tensorflow.keras.callbacks import Callback

class SaveWeights(Callback):
    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1)  % 5 == 0:
            self.model.save_weights(f'weights_at_epoch_{epoch + 1}.h5')

save_weights =  SaveWeights()
```

```python
# Train the model
nn.fit(X_train_scaled, y_train, epochs = 100, callbacks= save_weights)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
#Saving the model
nn.save('AlphabetSoupCharity.h5')
```
