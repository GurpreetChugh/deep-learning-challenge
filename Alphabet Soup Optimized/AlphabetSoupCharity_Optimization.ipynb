## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_reduced_df = application_df.drop(['EIN', 'NAME'], axis=1)
application_reduced_df.head()
```

```python
# Shape of the dataframe to ascertain number of records and features
application_reduced_df.shape
```

```python
# Determine the number of unique values in each column.
application_reduced_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_unique_counts = application_reduced_df['APPLICATION_TYPE'].value_counts()
app_type_unique_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace =  list(app_type_unique_counts[app_type_unique_counts < 528].index)
print(application_types_to_replace)

# Replace in dataframe
for app in application_types_to_replace:
    application_reduced_df['APPLICATION_TYPE'] = application_reduced_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_reduced_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_reduced_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts[classification_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(classification_counts[classification_counts < 1883].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_reduced_df.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_final_df = pd.get_dummies(application_reduced_df)
application_final_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_final_df.drop('IS_SUCCESSFUL', axis = 1).values
y = application_final_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

#### Hyperparameter tuning for Model Optimization


```python
from tensorflow.keras.optimizers import Adam

# Hyperparameter tuning
def create_model(hp):
    nn_model = Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation', ['relu','tanh'])

    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(Dense(units=hp.Int('first_units',
        min_value=32, max_value=512, step=32), activation=activation, input_dim=108))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int(f'units_{i}',
            min_value=32, max_value=512, step=32), activation=activation))

    nn_model.add(Dense(units=1, activation="sigmoid"))

    learning_rate = hp.Float("lr", min_value=1e-4, max_value=1e-2, sampling="log")

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer= Adam(learning_rate=learning_rate), metrics=["accuracy"])

    return nn_model
```

```python
import keras_tuner as kt
```

```python
#initializing the tuner
tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
#searching the best hyperparameters and model
tuner.search(X_train_scaled, y_train, epochs=20, validation_data=(X_test_scaled,y_test))
```

```python

#top hyperparameter
top_hyper = tuner.get_best_hyperparameters(1)[0]
top_hyper.values
```

```python
# top model
best_model = tuner.get_best_models(1)[0]
model_loss_best, model_accuracy_best = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss_best}, Accuracy: {round(model_accuracy_best*100, 2)}%")
```

```python
# saving the best model
best_model.save('AlphabetSoupCharity_Optimization.h5')
```
